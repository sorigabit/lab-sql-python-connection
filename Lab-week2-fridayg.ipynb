```python
def rentals_month(engine, month, year):
     data3 = pd.read_sql_query('SELECT *,month(rental_date) AS month_rental, year(rental_date) AS year_rental FROM sakila.rental', engine)
     df = pd.DataFrame(data3.query("`month_rental` == @month and `year_rental` == @year"))
     return df
```

```python

rentals_month(engine, 7, 2005)
```

```python
def rental_count_month(rentals_month, month, year):

        data4 = pd.DataFrame(rentals_month(engine, month, year).groupby(['customer_id'])['rental_id'].count())
        data4 = data4.rename(columns={"rental_id":f"rentals_{month}_{year}"})
        return data4
    
   
       
```
